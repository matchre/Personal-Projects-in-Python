```python
def fileconvert(filename):
    with open(filename) as content:
        file_data = content.read().splitlines() 
        
        h_p = []
        hh_p = []
        for i in range(len(file_data)):
            h_p = h_p + file_data[i].split('\n')
            hh_p = [rec.replace('"', "'") for rec in h_p]
    
    string = 'out.println("'
    back_string = '");'
    my_new_list = [string + x + back_string for x in hh_p]
    #print my_new_list
    print("\n".join(my_new_list))
    
    
    
fileconvert("filename.html")
```
